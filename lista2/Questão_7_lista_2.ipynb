7. Usando os dados temporais disponível no link a seguir sobre vendas de medicamentos na Austrália, que tipo de conclusões podemos obter em termos de tendência e sazonalidade? (2 pontos)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import matplotlib.dates as mdates
from datetime import date
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/bapimentel/Ciencia-de-Dados/master/Dados/Drugs.csv')

dados['date'] = pd.to_datetime(dados['date'])

dados
```

```python
dados.info()
```

```python
plt.figure(figsize=(23,12))
plt.plot(dados['date'], dados['value'])

plt.xticks(dados.date[5::4],size='small',rotation=35, ha="right")

plt.xlabel("Data", size = 16)
plt.ylabel("Valor", size = 16)
plt.title('Medicamentos vendidos na Austrália no período de 1991 e 2008')
```

Tendência não linear e sazonalidade multiplicativa
